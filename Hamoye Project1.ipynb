# Solution to Quiz



```python
# total number of unique countries in the dataset
unique_countries = df["Area"].nunique()
print("The total number of unique countries in the dataset is", unique_countries)
```

```python
# 7th lowest Area in 2017
grouped_by_area = df[['Area', 'Y2017']].groupby("Area").sum()
seventh_lowest = grouped_by_area.sort_values(by='Y2017').iloc[6]
print(seventh_lowest)
```

```python
# Area with highest sum value in 2017
grouped_by_area = df[['Area', 'Y2017']].groupby("Area").sum()
highest_area = grouped_by_area.sort_values(by='Y2017', ascending = False).iloc[0]
print(highest_area)
```

```python
# Total number of the sum of Processing in 2017
group_by_element_processing = df[df["Element"] == "Processing"]["Y2017"].sum()
group_by_element_processing
```

```python
# year with highest sum of Stock Variation
group_by_element_stock = df[df["Element"] == "Stock Variation"][["Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]].sum()
sum_of_stock_sorted = round(group_by_element_stock, 3).sort_values(ascending=False)
print(sum_of_stock_sorted.max())
sum_of_stock_sorted.plot(kind='bar', title='Sum of Stock Variation by Year')
plt.show()
```

```python
# Total number and percentage of missing data in 2014
Total_missing_data = df.isna().sum().sum()
Y2014_data = df['Y2014']
Total_missing_Y2014 = Y2014_data.isna().sum()
percentage_missing_Y2014 = (Total_missing_Y2014 / Total_missing_data) * 100
print('The total number of missing data in 2014 is ', Total_missing_Y2014)
print('The percentage of missing data in 2014 is', round(percentage_missing_Y2014, 3), '%')
```

```python
# year with least correlation with Element Code
correlation = df[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].corr()
correlation['Element Code'].idxmin()
```

```python
# Total sum of Wine produced in 2015 and 2018
group_by_item_wine = df[df["Item"] == "Wine"][["Y2015", "Y2018"]].sum()
group_by_item_wine
```

```python
# mean and standard deviation across the whole dataset for 2017
mean_2017 = df['Y2017'].mean()
std_2017 = df['Y2017'].std()
print(round(mean_2017,2))
print(round(std_2017,2))
```

```python
# Total Protein supply quantity in Madagascar in 2015
total_protein_2015 = df[(df['Area'] == 'Madagascar') & (df['Element'] == 'Protein supply quantity (g/capita/day)')]['Y2015'].sum()
total_protein_2015.round(2)
```

The Total Protein supply in Madagascar in 2015 is 173.05
